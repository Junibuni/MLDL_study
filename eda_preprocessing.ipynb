```python
import numpy as np
import pandas as pd
```

```python
data_path = './data/Loan_Default.csv'
num_sample = 10

missing_value_flag = 'remove'
scaling_flag = 'min-max'
```

```python
eda_df = pd.read_csv(data_path)
eda_df.head(num_sample)
```

```python
print(f'# of data: {eda_df.shape[0]}\n # of features: {eda_df.shape[1]}')

numerical_cnt = 0
categorical_cnt = 0

for feature_name, feature_type in zip(eda_df.columns, eda_df.dtypes):
    if np.issubdtype(feature_type, np.number):
        numerical_cnt += 1
    else:
        categorical_cnt += 1

    print(f'{feature_name}: {feature_type}')

print(f'# of Numerical Items: {numerical_cnt}\n # of Categorical Items: {categorical_cnt}')
```

```python
eda_df.describe()
```

```python
labeled_df = eda_df['Status']
label_dict = labeled_df.value_counts()

print(f'labeled shape: {labeled_df.shape}')
print(f'positive: {label_dict[0]/labeled_df.shape[0] * 100:.2f}%')
print(f'negative: {label_dict[1]/labeled_df.shape[0] * 100:.2f}%')
```

```python
duplicates = eda_df.duplicated()

for feature_name in eda_df.columns:
    if eda_df[feature_name].isnull().sum() > 0:
        eda_df[feature_name].fillna(eda_df[feature_name].mode().iloc[0], inplace=True)

outlier_dict = {}

for feature_name in eda_df.columns:
    if np.issubdtype(eda_df[feature_name].dtype, np.number):
        Q1 = eda_df[feature_name].quantile(0.25)
        Q3 = eda_df[feature_name].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        num_outlier = ((eda_df[feature_name] < lower_bound) | (eda_df[feature_name] > upper_bound)).sum()
        outlier_dict[feature_name] = num_outlier

        eda_df = eda_df[(eda_df[feature_name] >= lower_bound) & (eda_df[feature_name] <= upper_bound)]

print(f'# of outliers: {sum(outlier_dict.values())}')
for feature_name, num_outlier in outlier_dict.items():
    print(f'{feature_name}: {num_outlier}')
```
